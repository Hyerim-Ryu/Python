### array 연산

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]

list1 + list2
```

```python
list3 = []

for i in range(len(list1)):
    list3.append(list1[i] + list2[i])
list3
```

```python
# array는 기본적으로 요소별 연산을 진행한다.
arr1 = np.array(list1)
arr2 = np.array(list2)

print(arr1 + arr2)
print(arr1 * arr2)
```

### BMI 실습

```python
# np.loadtxt( "파일명.확장자" )
# delimiter : 구분기호
data = np.loadtxt( "height_weight.txt" , delimiter=',')
data
```

```python
# 1. data 에서 키와 몸무게를 나누기 !
height = data[ 0 ]
weight = data[ 1 ]
```

```python
# 2. 키(height 변수)를 cm -> m 단위로 바꾸기!
# 1cm = 0.01m
height_m = height * 0.01
```

```python
# 3. BMI 지수 구하기 (이미지 참고!)
bmi = weight / (height_m ** 2)
bmi
```

### 함수

#### 영화 평점데이터 연산

```python
# userId, itemId, rating, timestamp
ratings = np.loadtxt('ratings.txt', delimiter=',', dtype = np.int64)
ratings
```

```python
print(ratings.ndim)
print(ratings.shape)
print(ratings.size)
```

```python
# 전체 평점 평균 구하기
np.mean(ratings[ : , 2])
```

```python
# userId가 1인 user의 평점 평균 구하기
b = ratings[ : , 0] == 1
ratings[b, 2].mean()
```

```python
print(ratings.shape)
print(ratings[ : , 0 ].shape)
test = ratings[ : , 0 ] == 1
print(test.shape)
```

```python
# userId가 5인 user의 평점 평균 구하기
ratings[ratings[ : , 0] == 5, 2].mean()
```

```python
# 고유한 요소 1개씩 남겨서 배열 생성
unique_id = np.unique(ratings[ : , 0])

li = []
for user_id in unique_id :
    b = ratings[ : , 0 ] == user_id
    user_mean = ratings[b, 2].mean()
    li.append([user_id, user_mean])
```

```python
li[:5]
```

```python
ratings[ : , 0].shape
```

```python
ratings[ratings[ : , 1] == 1, 2].mean()
```

```python
# ItemId별 평점 평균 구하기

unique_item = np.unique(ratings[:, 1])

item_li = []
for item_id in unique_item : 
    b = ratings[ : , 1] == item_id
    item_mean = ratings[ b , 2 ].mean()
    item_li.append([item_id, item_mean])
item_li[ : 5 ]
```
