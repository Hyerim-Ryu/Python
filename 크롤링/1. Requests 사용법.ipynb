```python
# req -> 파이썬에서 브라우저의 역할을 대신
import requests as req
```

```python
# 1. req를 통해서 네이버 페이지 정보를 요청
# 2. response -> 응답코드를 넘겨받는다.
# 200, 300 -> 통신에 성공했습니다.
# 400 -> 클라이언트 요청의 문제가 있다.
# 500 -> 서버의 문제가 있다.
res = req.get( "http://www.naver.com" )
```

```python
res.text
```

# 뉴스타이틀 수집

```python
# 1. 네이버 사이트(검색한 뉴스화면) 를 요청
res = req.get("https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%EC%B4%88%EC%A0%84%EB%8F%84%EC%B2%B4")
res.text
```

# BeautifulSoup
- req를 통해서 받아온 텍스트 데이터를 컴퓨터가 이해할 수 있는 html형태로 변환

```python
# 주의할점. stone이 아닌 일반 soup을 import
from bs4 import BeautifulSoup as bs
```

```python
# 2. 응답을 받은 데이터에서 뉴스 타이틀만 수집
soup = bs(res.text, "lxml")
```

```python
# select("선택자") -> html 데이터에서 특정 요소만 수집
title = soup.select("a.news_tit")
```

```python
title
```

```python
# 수집된 요소 중에 컨텐츠(텍스트)데이터만 추출
title[0].text
```

```python
for i in title :
    print(i.text)
```

# 네이버금융정보를 수집

```python
# 1. req를 통해서 url 정보를 요청
finance = req.get("https://finance.naver.com/")
```

```python
# 2. 응답데이터가 올바르다면 bs을 통해서 html 형태로 변형
soup = bs(finance.text, "lxml")
```

```python
# 3. 변형된 데이터에서 필요한 태그만 수집
# 팁!부모자식을 의미하는 자식선택자를 잘 활용
# 수집하고자 하는 요소가 아이디, 클래스 구분자가 없다면 반드시 부모를 검사
# 부모태그가 구분자가 존재 할 때까지 검사를 진행
# 구분자가 없는 경우에는 정확하지 않을 확률이 높음
title = soup.select("#_topItems1 > tr > th > a ")
title
```

```python
for i in range(5) :
    print(title[i].text)
```

```python
for i in title :
    print(i.text)
```
