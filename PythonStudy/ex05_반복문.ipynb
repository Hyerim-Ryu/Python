#### 반복문


- while문

```python
while True :
    num1 = int(input('첫 번째 정수 입력 >> '))
    num2 = int(input('두 번째 정수 입력 >> '))
    
    
    if num1 == 0 and num2 == 0 :
        print('프로그램 종료')
        break;
    else : 
        print(f'두 정수의 합 : {num1+num2}')
```

```python
num1, num2 = 0,1
while num1!=0 or num2!=0 :
    num1 = int(input('첫 번째 정수 입력 >> '))
    num2 = int(input('두 번째 정수 입력 >> '))  
    
    if num1 != 0 or num2 != 0 :
        print(f'두 정수의 합 : {num1+num2}')       
```

```python
import random

randnum = random.randint(1,50)

while True :
    num = int(input('숫자를 입력하세요 >> '))
    if randnum > num :
        print(f'{num}보다 큰 수 입니다.')
    elif randnum < num :
        print(f'{num}보다 작은 수 입니다.')
    else :
        print('정답을 맞추셨습니다.')
        break
```

#### 리뷰 길이 평균 값


```python
total = 0
for i in range(len(movie_review)) :
    total += (len(movie_review[i][0]))
print(f"리뷰의 평균 길이는 '{total/len(movie_review)}'입니다.")
```

```python
total = 0
for txt, pn in movie_review :
    total += len(txt)
print(f"리뷰의 평균 길이는 '{total/len(movie_review)}'입니다.")
```

#### 긍정리뷰/부정리뷰 개수


```python
p = 0
n = 0
for txt,pn in movie_review :
    if pn == 0 :
        n += 1 
    else :
        p += 1
print(f"긍정 '{p}'개\n부정 '{n}'개")
```

```python
pnLi = []
for txt, pn in movie_review :
    pnLi.append(pn)
print(f"긍정 '{pnLi.count(1)}'개")
print(f"부정 '{pnLi.count(0)}'개")
```

#### 리뷰길이 최대 값 / 리뷰 내용 출력하기

```python
max = 0
for txt,pn in movie_review :
    if max < len(txt) :
        max = len(txt)
        text = txt
print(f"리뷰 최대길이는 '{max}'입니다.\n{text}")
```

```python
text = input('검색할 데이터 입력 >> ')
reCnt = 0

for txt,pn in movie_review :
    if text in txt :
        reCnt+=1
print(f"'{text}'가 포함된 리뷰는 {reCnt}개 입니다.")
```
